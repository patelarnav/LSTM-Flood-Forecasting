```python
import numpy as np
import pandas as pd
import os
```

```python
data = pd.read_excel('data\Observed_discharge_upto_basantpur_2000_2014.xlsx')
```

```python
start_date = pd.to_datetime('2001-06-01')
end_date = pd.to_datetime('2001-09-30')

# Iterate over each date and create separate atables
current_date = start_date
while current_date <= end_date:
    filtered_data = data[data['Date'].dt.month == current_date.month][data['Date'].dt.day == current_date.day]
    filtered_data = filtered_data.sort_values('Date')  # Sort the data by date

     # Generate the table name
    table_name = f"{current_date.strftime('%m-%d')}_data.csv"
    
    # Define the folder path
    folder_path = "n_data"
    
    # Create the folder if it doesn't exist
    os.makedirs(folder_path, exist_ok=True)

    # Save the filtered data to the table in the specified folder
    file_path = os.path.join(folder_path, table_name)
    filtered_data.to_csv(file_path, index=False)
    # Move to the next date
    current_date += pd.DateOffset(days=1)
```
